```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv(r"C:\Users\HP-PC\Desktop\Python data analytics\mail_data.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data['Category'] = data['Category'].replace({'spam': 0, 'ham': 1})
```

```python
data
```

```python
data.drop('category', axis=1, inplace=True)
```

```python
data
```

```python
x = data['Message']
y = data['Category']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
x_test.shape
```

```python
extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
```

```python
x_train_features=extraction.fit_transform(x_train)
x_test_features=extraction.transform(x_test)
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
print(x_train_features)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_features,y_train)
```

```python
prediction_train_data=model.predict(x_train_features)
accuracy_train_data=accuracy_score(y_train,prediction_train_data)
```

```python
print(accuracy_data)
```

```python
prediction_test_data=model.predict(x_test_features)
accuracy_test_data=accuracy_score(y_test,prediction_test_data)
```

```python
print(accuracy_test_data)
```
